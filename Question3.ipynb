###A palindrome is a case-insensitive word, phrase, or sequence that reads the same backwards as forwards.

###Write a program that checks whether a given string is a palindrome.

```python
string = input()

new_string = ""

reverse_string = ""


#This code removes all the panctuation and turns all letters into lower case
for i  in range(len(string)):

    if string[i] in "_ -',.?!+<>@#$%^&*()[]/~;:":
        continue
    else:
        new_string += string[i].lower()


#this code reverses the new string that was created
#for i in range(len(new_string)-1,-1,-1):
#    reverse_string += new_string[i]


reverse_string = new_string[::-1] #the best way is to reverse the string using string slicing
print(reverse_string)

#this condition checks weather the the reverse string and the new_string are the same
if new_string == reverse_string:
    print(True)
else:
    print(False)
```
